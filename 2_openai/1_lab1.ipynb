```python
# The imports

from dotenv import load_dotenv
from agents import Agent, Runner, trace, OpenAIChatCompletionsModel, set_tracing_disabled
from openai import AsyncOpenAI

import os
```

```python
# The usual starting point
load_dotenv(override=True)

url = os.getenv("URL", "https://integrate.api.nvidia.com/v1")
print(f"Using NVIDIA API URL: {url}")


api_key = os.getenv("NVIDIA_API_KEY")

MODEL_NAME="microsoft/phi-3-small-128k-instruct"

client = AsyncOpenAI(
  base_url = "https://integrate.api.nvidia.com/v1",
  api_key = api_key
)
set_tracing_disabled(disabled=True)
```

```python

# Make an agent with name, instructions, model

# agent = Agent(name="", instructions="", model="gpt-4o-mini")

agent = Agent(
        name="Jokester",
        instructions="You are a joke teller",
        model=OpenAIChatCompletionsModel(model=MODEL_NAME, openai_client=client)
    )
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
```
